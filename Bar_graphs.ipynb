### Bar Graphs

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# a basic bar graph

phone_brand = ['Apple', 'Sumsung', 'Nokia', 'Oppo', 'Redmi']
sales = [120, 150, 20, 50, 80]

plt.bar(phone_brand, sales)
plt.show()
```

### utilizing the various parameters

```python
# Example data
clubs = ['Science', 'Drama', 'Art', 'Music', 'Sports']
students = [45, 30, 25, 40, 50]

# Create bar graph
plt.bar(clubs, students,        # We start with the categorical data as the x arguements
        color= 'yellowgreen',   # Color of the bars
        edgecolor= 'black',     # color of the line used to draw the bars
        linewidth= 3,           # thickness of the line used to draw the bars   
        width= 0.7              # relative size ratio of the bars
        #align= 'edge'          # x label position relative to the bars
       )

# Labeling and styling
plt.title('Number of Students in School Clubs')
plt.xlabel('Clubs')
plt.ylabel('Number of Students')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Look at this work of art

```python
clubs= ['sports', 'art', 'music', 'drama', 'cycling', 'sci', 'journalism', 'chess']
population = [100, 80, 150, 300, 400, 240, 50, 20]


plt.bar(clubs, population, 
        color='skyblue',
        edgecolor= 'black',
        linewidth= 5,
        width= 0.7
)
plt.title('Members population of the clubs in a school')
plt.grid(axis= 'y', linestyle= '--' , alpha= 0.7)

plt.show()
```

### Using the same information, we can sort the bar graph in ascedning order using the zip and sorted function

```python
sorted_pairs = sorted(zip(clubs, population), key=lambda x: x[1])
sorted_clubs, sorted_population = zip(*sorted_pairs)

# plt.figure(figsize=(10, 5))
plt.bar(sorted_clubs, sorted_population,
       color='skyblue',
       edgecolor= 'black',
       linewidth= 5,
        width= 0.7)

plt.title('Club Populations (Ascending Order)')
plt.xlabel('Clubs')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
