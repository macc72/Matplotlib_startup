# Histograms

```python
import numpy as np
from matplotlib import pyplot as plt
```

### a simple histogram


```python
# Sample data
data = [5, 7, 8, 5, 6, 9, 7, 6, 5, 8, 9, 7, 6, 6, 8]

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')

# Show plot
plt.show()
```

### A more beautiful histogram :)

```python
# Example data: Random exam scores for 100 students
np.random.seed(42)
scores = np.random.normal(70, 10, 100)  # mean=70, std=10

# Create histogram
plt.hist(scores,
    bins=10,                  # Number of bins
    color='skyblue',          # Fill color
    edgecolor='black',        # Border color
    alpha=0.8,                # Transparency
    density=False,            # Show frequency (False) or probability density (True)
    histtype='bar',           # Type: 'bar', 'barstacked', 'step', or 'stepfilled'
    orientation='vertical',   # 'vertical' or 'horizontal'
    rwidth=0.9,               # Width of bars relative to bin width (01)
    label='Exam Scores'       # For use with legend
  # cumulative= True          # Setting the cumulative as true does the cumulative of the scores
)

# Add labels and title
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam Scores')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

### Another detailed example

```python
data = np.random.normal(30, 30, 100)

# Create histogram
plt.hist(data,
        bins= 30,
        rwidth= 0.9,
        color='skyblue',
        label='Normal Dist',
        range= (0, 100) 
        )

# Add legend and labels
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.title('Histogram with All Parameters')
plt.legend()
plt.grid(True)

plt.show()
```

### We can also make a histogram generating numbers using np.random.normal() function 

```python
data = np.random.normal(loc=10, scale=2, size=10000)

plt.hist(data, bins=50, color='skyblue',edgecolor='black')

plt.axvline(10, color='red', label='Mean (10)')
plt.axvline(8, color='green', linestyle='--', label='1 SD below (8)')
plt.axvline(12, color='green', linestyle='--', label='1 SD above (12)')

plt.title("Normal Distribution: mean=10, std=2")
plt.legend()
plt.show()
```
