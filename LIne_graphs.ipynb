# Line graphs

```python
# importing the necessary libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
# a basic line graph

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.show()
```

### - A broad exampe where we utilise most of the parameters

```python
# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Plotting the line graph
plt.plot(x, y,  
         # Line paramters
         color='green', # line color
         linestyle='-', # Line style: solid (-), dashed (--), dashdot (-.), dotted (:)        
         linewidth=2,   # Line width 
         
         # Marker parameters 
         marker= 'o',             # Marker for data points
         markersize= 8,           # Marker size
         markerfacecolor= 'blue', # Marker fill color
         markeredgecolor= 'black',# Marker edge color
         markeredgewidth= 1,      # Thinckness of the edge of the marker
         label = "y = x^2"
        )

# Adding titles and labels
plt.title("Square Numbers Line Graph")

plt.xlabel("X values")
plt.ylabel("Y = X^2")

# Grid and legend
plt.grid(True)      # Show grid
plt.legend()        # Add legend

# Show the plot
plt.show()
```

### - We can also plot multiple lines on a single graph

```python
# X values (shared)
x = [0, 1, 2, 3, 4, 5]

# Different Y values
y1 = [0, 1, 4, 9, 16, 25]     # y = x^2
y2 = [0, 1, 2, 3, 4, 5]       # y = x
y3 = [0, 1, 8, 27, 64, 125]   # y = x^3

# Plot each line with custom styles
plt.plot(x, y1, label="y = x^2", color='green', linestyle='--', marker='o')
plt.plot(x, y2, label="y = x", color='blue', linestyle='-', marker='s')
plt.plot(x, y3, label="y = x^3", color='red', linestyle='-.', marker='^')

# Add labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Multiple Line Graphs")

# Show legend and grid
plt.legend()  # the legend for each line has already been specified for each line as a parameter
plt.grid(True)

# Display the graph
plt.show()
```

### - If the x-values are not common to each line

```python
# Line 1: Quadratic, with its own x values
x1 = [0, 1, 2, 3, 4]
y1 = [200, 100, 150, 290, 300]

# Line 2: Linear, different x range
x2 = [2, 3, 4, 5, 6]
y2 = [250, 200, 100, 80, 50]

# Line 3: Cubic, different and non-aligned x values
x3 = [1, 3, 5, 7]
y3 = [1, 27, 125, 343]

# Plot each line separately with its own x and y
plt.plot(x1, y1, label="Foirella", color='green', marker='o')
plt.plot(x2, y2, label="Coops", color='blue', marker='s')
plt.plot(x3, y3, label="Cambium", color='red', marker='^')

# Labeling and styling
plt.title("Multiple Lines with Independent X Points")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)

plt.show()
```

### We can make dotted line graphs

```python
x = np.arange(0, 11, 0.5) # Here we use arange to make the line smooth as we have small steps of 0.5
p =  int(0.6 * len(x)) # 0.6 = 60% of the graph solid

plt.plot(x[:p], x[:p]**2, 'b-')
plt.plot(x[p-1:], x[p-1:]**2, 'r:')
plt.show()
```

### Numpy offers trigonometric functions that we can utilize in our line graphs

```python
# Example data
x = np.linspace(1, 15, 100) # similarly: x = np.arange(1, 5, 0.05)
y = np.sin(x)

# Split point index (e.g., 50% solid, 50% dotted)
split_idx = int(len(x) * 0.5)

# Plot solid part
plt.plot(x[:split_idx], y[:split_idx], 'b-', label='Solid start')

# Plot dotted part
plt.plot(x[split_idx-1:], y[split_idx-1:], linestyle=':', color='blue', label='Dotted end')

plt.title("Line transitioning from solid to dotted")
plt.legend()
plt.grid(True)
plt.show()
```
