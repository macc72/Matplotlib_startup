# Pie charts

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# a simple pie chart

langs = ['Python', 'Java', 'C#', 'Javascript', 'C++']
votes = [60, 20, 30, 55, 35]
 
plt.pie(votes, labels= langs)
plt.show()
```

```python
# same pie with modifiacitons

langs = ['Python', 'Java', 'C#', 'Javascript', 'C++']
votes = [60, 20, 30, 55, 35]  
explodes= [0, 0.3, 0, 0, 0]
 
plt.pie(votes, labels= langs,  # instead of x and y in pie chart we have; amount and categories(labels)  
        explode= explodes,     # this pulls out the segment from the center 
        autopct= "%.2f%%",     # I still don't get this format but I roll with it
        pctdistance= 1.3,      # distanec of percentages from the middle
        labeldistance= 0.4,    # distanec of labels from the middle
        startangle= 90,        # what angle to start the first segment
        counterclock=False    # self explainatory :) 
       )
plt.show()
```

### One more example is not too much to ask for

```python
# Example data
labels = ['Science', 'Math', 'Art', 'Sports']
sizes = [25, 30, 20, 25]
colors = ['gold', 'skyblue', 'lightcoral', 'yellowgreen'] # got some creative colors from gpt
explode = (0.1, 0, 0, 0)  # 'explode' the 1st slice

# Create pie chart
plt.pie(
    sizes,              # Data values
    labels=labels,      # Labels for each slice
    colors=colors,      # Slice colors
    explode=explode,    # Offset one slice
    autopct='%1.1f%%',  # Display percentage text
    shadow=True,        # Add shadow effect
    startangle=90,      # Rotate start angle
    counterclock=False, # Draw clockwise
    wedgeprops={'edgecolor': 'black'},# Border color
    labeldistance=1.1   # Distance of labels from center
)
# wedgeprops={'width': 0.4} for creating donut chart 

# Add title
plt.title('Distribution of Club Participation')

# Equal aspect ratio ensures pie is circular
plt.axis('equal')

# Display the pie chart
plt.tight_layout()
plt.show()
```
